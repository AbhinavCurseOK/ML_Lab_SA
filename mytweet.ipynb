```python
import tweepy
consumer_key = '#'
consumer_secret = '#'
bearer_token = '##'
client = tweepy.Client(bearer_token=bearer_token)
```

```python
import pandas as pd
import os
```

```python
response = client.search_recent_tweets(query='Blockchain Security', max_results=10)
if response.data:
    tweets_list = [tweet.text for tweet in response.data]
    new_tweets_df = pd.DataFrame(tweets_list, columns=['Tweet'])

    file_path = 'tweets.csv'
    if os.path.exists(file_path):
        new_tweets_df.to_csv(file_path, mode='a', header=False, index=False)
    else:
        new_tweets_df.to_csv(file_path, index=False)

    print("Tweets saved to 'tweets.csv'.")
else:
    print("No tweets found.")
```
